```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Hiển thị đồ thị ngay trong notebook
%matplotlib inline

# Thiết lập style cho biểu đồ
sns.set(style="whitegrid")
```

```python
# Đọc dữ liệu từ file csv
df = pd.read_csv('../Vu_data/data.csv')

# Hiển thị thông tin cơ bản
print(df.info())
df.head()
```

```python
# Loại bỏ cột 'Unnamed: 32' (lỗi thường gặp) và cột 'id'
if 'Unnamed: 32' in df.columns:
    df = df.drop(columns=['Unnamed: 32'])
if 'id' in df.columns:
    df = df.drop(columns=['id'])

print("Dữ liệu sau khi xử lý:")
df.head()
```

```python
plt.figure(figsize=(8, 6))
sns.countplot(x='diagnosis', data=df, palette='viridis')
plt.title('Phân phối của chẩn đoán (Diagnosis)')
plt.xlabel('Diagnosis (M = Malignant, B = Benign)')
plt.ylabel('Số lượng')
plt.show()
```

Mục đích: Biểu đồ này đếm số lượng các mẫu bệnh phẩm cho từng loại chẩn đoán: Lành tính (Benign - B) và Ác tính (Malignant - M).

Ý nghĩa trong bài toán này:

Giúp chúng ta thấy được sự cân bằng của dữ liệu.

Thông thường, số lượng ca lành tính (B) sẽ nhiều hơn ác tính (M). Nếu sự chênh lệch quá lớn (ví dụ: 90% B và 10% M), dữ liệu bị coi là mất cân bằng (imbalanced), và bạn sẽ cần lưu ý khi chọn độ đo đánh giá mô hình (không nên chỉ dùng Accuracy mà nên dùng F1-score hoặc ROC-AUC).

```python
# Lọc các cột chứa chữ "_mean"
mean_features = [col for col in df.columns if '_mean' in col]

plt.figure(figsize=(12, 10))
corr_matrix = df[mean_features].corr()
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title('Ma trận tương quan giữa các đặc trưng trung bình (Mean Features)')
plt.show()
```

Mục đích: Thể hiện mối quan hệ tuyến tính giữa các đặc trưng (features) với nhau. Màu càng đậm (đỏ/nóng) thì tương quan càng cao (gần 1), màu càng lạnh (xanh) thì tương quan càng thấp hoặc âm.

Ý nghĩa trong bài toán này:

Bạn sẽ thấy các cụm biến có màu đỏ rực như radius_mean (bán kính), perimeter_mean (chu vi) và area_mean (diện tích). Điều này dễ hiểu vì về mặt hình học, bán kính càng lớn thì chu vi và diện tích càng lớn.

Cảnh báo: Hiện tượng này gọi là đa cộng tuyến (multicollinearity). Khi các biến quá giống nhau, chúng cung cấp thông tin trùng lặp cho mô hình máy học. Trong các bước nâng cao, người ta thường loại bỏ bớt (ví dụ: chỉ giữ lại radius, bỏ perimeter và area) để mô hình chạy nhanh hơn và ổn định hơn.

```python
features_to_plot = ['radius_mean', 'texture_mean', 'area_mean', 'smoothness_mean']

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
axes = axes.flatten()

for i, feature in enumerate(features_to_plot):
    sns.histplot(data=df, x=feature, hue='diagnosis', kde=True, element="step", ax=axes[i], palette='viridis')
    axes[i].set_title(f'Phân phối của {feature} theo Diagnosis')

plt.tight_layout()
plt.show()
```

Mục đích: So sánh hình dáng phân phối dữ liệu của từng đặc trưng đối với hai nhóm bệnh (Lành tính vs Ác tính).

Ý nghĩa trong bài toán này:

Khả năng phân loại:

Nếu hai đường phân phối (màu xanh lá và màu xanh) tách biệt nhau rõ ràng (ví dụ: radius_mean, area_mean), nghĩa là đặc trưng đó rất tốt để phân biệt bệnh. Người bị ung thư ác tính thường có kích thước khối u lớn hơn hẳn.

Nếu hai đường này chồng lấn lên nhau (ví dụ: smoothness_mean, texture_mean), nghĩa là đặc trưng đó khó dùng để phân biệt riêng lẻ. Cả u lành và u ác đều có độ mịn hoặc kết cấu tương tự nhau.

```python
plt.figure(figsize=(14, 8))
data_melted = pd.melt(df, id_vars="diagnosis", value_vars=['radius_mean', 'texture_mean', 'perimeter_mean', 'compactness_mean'],
                      var_name="features", value_name="value")
sns.boxplot(x="features", y="value", hue="diagnosis", data=data_melted, palette='viridis')
plt.xticks(rotation=45)
plt.title('So sánh các đặc trưng trung bình theo Diagnosis')
plt.show()
```

Mục đích: Tóm tắt các thông số thống kê (trung vị, tứ phân vị) và phát hiện các giá trị ngoại lai (outliers).

Ý nghĩa trong bài toán này:

Hộp (Box): Đường gạch ngang trong hộp là giá trị trung vị (median). Bạn sẽ thấy hộp của nhóm Ác tính (M) thường nằm cao hơn nhóm Lành tính (B) ở các đặc trưng về kích thước.

Điểm đen (Outliers): Các chấm đen nằm ngoài hai đầu "râu" của biểu đồ là các giá trị ngoại lai. Điều này cho thấy có một số ca bệnh có chỉ số cực kỳ cao bất thường so với mặt bằng chung.

Biểu đồ này xác nhận lại nhận định từ Histogram một cách cô đọng hơn: Nhóm ác tính có xu hướng có các chỉ số trung bình cao hơn và độ biến động (chiều dài hộp) lớn hơn.
